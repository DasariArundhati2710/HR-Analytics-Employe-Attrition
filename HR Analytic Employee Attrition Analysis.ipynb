```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Attrition", data=df)
plt.title("Employee Attrition Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Department", hue="Attrition", data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MonthlyIncome"], bins=30, kde=True)
plt.title("Monthly Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="JobSatisfaction", hue="Attrition", data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="OverTime", hue="Attrition", data=df)
plt.title("Overtime Impact on Attrition")
plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
avg_income = df.groupby("JobRole")["MonthlyIncome"].mean().sort_values()

plt.figure(figsize=(10,6))
avg_income.plot(kind="barh")

plt.title("Average Monthly Income by Job Role")
plt.xlabel("Monthly Income")
plt.show()
```

### Key Insights

• Employees working overtime show higher attrition rates.

• Some departments have higher employee turnover.

• Monthly income and job satisfaction influence employee retention.

• Work-life balance may impact employee attrition.
